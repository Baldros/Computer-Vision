````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````



```

```



````python
#Função para plotar as imagens:
def display_img(imagem,cmap=None):
  fig=plt.figure(figsize=(8,6))
  ax = fig.add_subplot(111)
  ax.imshow(imagem, cmap=cmap)
````

````python
moedas = cv2.imread('moedas.jpg')
````

````python
display_img(moedas)
````



```
É de boas práticas aplicar um suavisação.

A suavização utilizada foi a blur de mediano.
```



````python
#Aplicando suavização:
gray_moedas = cv2.cvtColor(cv2.medianBlur(moedas, ksize=25),cv2.COLOR_BGR2GRAY)
````



```
Note que tem como fazer todo o processo de conversões de uma única vez,
logo, não há a necessidade de ficar fazendo demoradamente parte por parte.

Joga um código dentro do outro que é GG... 😉
```



````python
display_img(gray_moedas, cmap='gray')
````



```
Falta ainda uma ultima etapa, a aplicação do Threshold.
```



````python
#Criando a imagem Thresh:
ret, moedas_thresh = cv2.threshold(gray_moedas,160,255,cv2.THRESH_BINARY_INV)
````

````python
display_img(moedas_thresh, cmap='gray')
````



```
A ideia é achar os contornos na imagem simplificada, para depois achar os
contornos na imagem final.
```



````python
kernel = np.ones((3,3),np.uint8)
opening = cv2.morphologyEx(moedas_thresh,cv2.MORPH_OPEN, kernel, iterations = 2)
````

````python
dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2,5)
````
