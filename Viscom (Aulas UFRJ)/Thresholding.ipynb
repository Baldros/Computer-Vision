

```
Tresholding, Binarização, ou Lineariação nada mais é que a conversão
dos bits da imagem para binário, ou zero ou um.
```



#Importações

````python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
````

#Código

## Importando a imagem

````python
#Importando a Imagem:
img = cv2.imread('Steve Vai.jpg', 0)
````

````python
#Plotando a imagem:
plt.imshow(img, cmap = 'gray')
````



```
Para fazer a binarização, o ideal é pegar o "Suit Spot" que nada mais é que
o valor médio. Por isso é interessante saber o valor máximo e mínimo.
```



````python
#Chegando máximo e m
print(img.max())
print(img.min())
````



```
Na maioria dos casos, o valor máximo e mínimo serão esses encontrados, porém,
pode haver algum caso particular.
```



## Binarizando

````python
#Binarizando:
ret, img_thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
````



```
Note que essa função do cv2 recebe de retorno dois argumentos, ou seja,
há a criação de duas variaveis para uma única função. Isso pode parecer
estranho a primeira vista.

ret -> o próprio threshold
img_thresh -> imagem com o threshold aplicado

OBS: O nome das variaveis podem ser quaisquer... Mas o retorno é nessa ordem.
```



````python
#plotando a imagem:
plt.imshow(img_thresh, cmap='gray')
````

````python
#Mostrando o Array
img_thresh
````



```
Essa imagem não mostra, mas os pontos brancos estão como 255 e não 
como 1, que é como deveria estar numa binarização.

Existem algumas técnicas para resolver essa questão, dentre elas, converter o
array para float, ou simplesmente dividir o array por 255 (ou o valor máximo
que constar nele).

OBS: O ideal desse segundo metodo, que é mais um macete, é utilizar outra imagem
para não correr o risco de bagunçar tudo.
```
