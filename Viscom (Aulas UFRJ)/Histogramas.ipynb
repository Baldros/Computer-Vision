#importações

````python
import  cv2 #Para ler e mexer na imagem;
import numpy as np #Para mexer no array, se preciso for;
import matplotlib.pyplot as plt #Para plotar a imagem, já que uma imagem é um array.
%matplotlib inline
````

#Importando a imagem:

````python
#importando a imagem:
img = cv2.imread('Batman.jpg')
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
````



```
Lembre que o cv2 importa a imagem como BGR, sendo assim
há a necessidade de converter essa imagem para RGB.
```



````python
#Plotando a imagem
plt.imshow(img_RGB)
````

#Construindo os Histogramas

### Construindo um histograma simples:



```
A plotagem do histograma serve para checar a intensidade da cor
de forma mais visual.

Daria para ver isso pelo array? Provavelmente... Maaas,
o histograma deixa a coisa mais visual.
```



````python
#Construindo um histograma:
hist_red = cv2.calcHist([img],channels = [2], mask=None,histSize=[526], ranges=[0,255])
````



```
Note que a imagem utilizada foi a inicial, em BGR. O channel (Canal)
respeitará esta ordem, ou seja: Imagem tá me BGR?
channels = [0] -> Blue
channels = [1] -> Green
channels = [2] -> Red
```



````python
#Plotando o Gráfico:
plt.plot(hist_red, color = 'red')
````

## Construindo um histograma multiplo



```
A ideia aqui é utilizar um for para plotar o histograma de cada
canal na mesma imagem.

OBS: A imagem já foi plotada desde o começo, e continuará sendo utilizada
a imagem em BGR:
```



````python
#Plotando o histograma multiplo:
color = ('b', 'g', 'r') #Note que a ordem foi mantida!
for i, col in enumerate(color):
  histr = cv2.calcHist([img],[i],None,[256],[50,200])
  plt.plot(histr, color = col)
  plt.xlim([0,256])
plt.show()
````



```
Note que o codigo de plotagem do Histograma é o mesmo da celula do 
hist_red, a diferença é que, como o professor escreveu todos os parâmetros na
ordem padrão, não precisou definir o que é o que.
```
