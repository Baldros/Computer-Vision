#vetorização e plotagem

Importando a imagem com o PIL e plotando com o matplotlib

1.   Converção - image PIL para um array numpy;
2.   Plotagem - Utilizando o matplotlib para plotar o array como imagem.



---

*   É necessário passar a imagem para array, para poder fazer as alterações necessárias nela.

*   Devido ao fato dela passar a ser um array, há a necessidade de utilizar 
uma outra forma de mostrar, que é a plotagem, através do matplotlib.




````python
#importando imagem:
img = Image.open('webb.jpg')
display(img)
print('O tipo da imagem é de {}'.format(type(img)))
````

````python
#Convertendo imagem em array:
img_arr = np.asarray(img) 
display(img_arr)
print('\n')
print('O tipo da imagem é da {}\n'.format(type(img_arr)))
print('O shape dessa imagem é {}'.format(img_arr.shape))
````

#Opencv



1.   Importação:
2.   Correções:







```
O Opencv é uma outra forma de se carregar a imagem, porém ele já abre 
a imagem como um array, o que faz trás a necessidade do numpy.

Note a importância do numpy quando se fala em tratamento de arrays.
```





````python
#importando a imagem:
img_cv = cv2. imread('webb.jpg')
print('A tipo do arquivo é de classe {}'.format(type(img_cv))) #para ver o tipo
print(img_cv.shape) #para ver as caracteristicas
````



```
Porém é valido pontuar uma coisa sobre o Opencv...
```



````python
plt.imshow(img_cv) #Note que o matplotlib continua sendo usado para plotar as imagens
````



```
Por padrão, o opencv abre as imagens em BGR, ou seja, ele reorganiza 
a ordenação das camadas, sendo assim, pode ser que se seja necessário 
que a imagem seja convertida para RGB.

Ou não, pode ser que esse fator seja irrelevante, ou até mesmo preferivel.

Preferivel porque o formato BGR nasce justamente de uma necessidade, a limtação
computacional. Uma necessidade computacional, em geral, já superada, porém o 
formato tá ai até os dias atuais e continua sendo majoritáriamente usado.
```



#Converções:

Converção de BGR para RGB



```
Existem várias tipos de conversão, no caso usar-se-ara aqui o tipo "BGR2RGB".

Vale ressaltar aqui também que os tipos de conversão são, por padrão todos MAIÚSCULOS.
```




````python
#Convertendo
img_rgb = cv2.cvtColor(img_cv,cv2.COLOR_BGR2RGB)

#Plotando
plt.imshow(img_rgb)
````

Converção para gray



```
Uma forma de fazer essa conversão, é simplesmente não convertendo, mas sim já
importando a imagem com apenas um canal de cor...
```



````python
img_gray = cv2.imread('webb.jpg',cv2.IMREAD_GRAYSCALE)
#note a diferenta
print('O shape do img_gray é {}, enquanto o shape da img_rgb é {}'.format(img_gray.shape,img_rgb.shape))
````



```
Note que o shape da img_gray não informa os canais de cores... Isso, justamente
porque ela veio apenas com um canal, o canal cinza.
```





```
Porém, note...
```



````python
plt.imshow(img_gray)
````



```
Ou seja, ainda que se importe uma imagem com com apenas um canal, o opencv, por
padrão, sempre vai plotar a imagem nessa escala "viridis"

Ou seja, para plotar em cinza, ainda que tennha sido lida em escada de cinza...
precisa colocar o cmpa = "gray" 🙄

Assim sendo...
```



````python
plt.imshow(img_gray, cmap = "gray")
````

Calculando máximos e mínimos

````python
#Máximo
print(img_gray.max())

#minimo
print(img_gray.min())
````



```
Note que o valor máximo não veio o 255 que é o valor máximo teórico. Isso pode
acontecer... Ou seja, esteja acostumado com essa possibilidade de vir um valor
anterior ao extremo esperado.

Isso serve pro minimo também, pode ser que o minimo seja diferente de zero... 
Não é anormal de se acontecer.
```



#Rescalonamento da Imagem

 Função resize



```
Escalonamento da imagem
```

````python
img_res = cv2.resize(img_rgb,(0,0), img_rgb,2,2)
#para ser proporcional, os valores precisam ser igual (Os ultimos dois valores) 
plt.imshow(img_res)
````
