#Importações

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

#Código

##Preparação

````python
def display_img(imagem):
  fig=plt.figure(figsize=(8,6))
  ax = fig.add_subplot(111)
  ax.imshow(imagem,cmap='gray')
````



```
Para facilitar evitar toda hora ficar se repetindo, como serão plotadas algumas
imagens, então é mais facil definir uma função para o caso.
```



##Importando a imagem

````python
#Importando a imagem
img = cv2.imread('nebulosa.jpg', 0)
````

````python
#Checando o Shape
img.shape
````



```
O comando shape serve para dar a altura e a largura da imagem, respectivamente.
```



##Aplicando o 1º tipo de Sobel (Gradiente) - Sobel X

````python
#Criando a imagem Sobel X:
img_sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
````

````python
#plotando a imagem:
display_img(img_sobelx)
````

##Aplicando o 2º tipo de Sobel (Gradiente) - Sobel Y

````python
#Criando a imagem Sobel Y:
img_sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
````

````python
#plotando a imagem:
display_img(img_sobely)
````

##Conclusão:
SobelX vs SobelY



```
Note que na verdade, a diferença entre o SobelX para o SobelY é nítida e se
relaciona com os elementos da imagem. Se é SobelX os elementos afundam na
imagem, se o sobelY os elementos se sobressaem. É algo bem intuitivo.
```



##Aplicando o 3º tipo de Sobel (Gradiente) - Sobel addWeighted

````python
#Criando a imagem Sobel addWeighted:
img_blendedsb = cv2.addWeighted(src1=img_sobelx,alpha=0.5,
                                src2=img_sobely,beta=0.5,
                                gamma=0)
````



```
Note que para a criação desse Sobel, precisamos criar primeiro as imagens
gradientes em relação a X e em relação a Y.
```



````python
display_img(img_blendedsb)
````

````python
display_img(img_laplace)
````
